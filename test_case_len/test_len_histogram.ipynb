```python
import matplotlib.pyplot as plt
import pandas as pd


def make_plot(len_df, save_filename):
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    ax.bar(len_df['mig_name'],len_df['sys'], bottom =len_df['oracles']+len_df['gui'],  color='tab:green', label='Sys event')
    ax.bar(len_df['mig_name'],len_df['oracles'], bottom=len_df['gui'], color='tab:orange', label= 'Oracles events')
    ax.bar(len_df['mig_name'],len_df['gui'], color='tab:blue',label='GUI events')

    # ax.bar(len_df['mig_name'],len_df['gui'],label='GUI events')
    # ax.bar(len_df['mig_name'],len_df['oracles'], label= 'Oracles events')
    # ax.bar(len_df['mig_name'],len_df['sys'],label='Sys event')


    plt.xticks(rotation = 90)
    plt.legend(loc=1)
    # plt.show()
    plt.savefig(save_filename, bbox_inches= 'tight')

targets_len = pd.read_csv('target_test_len.csv')
targets_len_craft = targets_len[targets_len['mig_name'].str.contains('a[1-5]')]
targets_len_atm = targets_len[~targets_len['mig_name'].str.contains('a[1-5]')]
make_plot(targets_len_craft, 'target_craft_len.pdf')
make_plot(targets_len_atm, 'target_atm_len.pdf')


targets_len = pd.read_csv('src_test_len.csv')
targets_len_craft = targets_len[targets_len['mig_name'].str.contains('a[1-5]')]
targets_len_atm = targets_len[~targets_len['mig_name'].str.contains('a[1-5]')]
make_plot(targets_len_craft, 'src_craft_len.pdf')
make_plot(targets_len_atm, 'src_atm_len.pdf')
```
